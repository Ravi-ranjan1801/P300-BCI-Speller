This notebook handles raw dataset loading and validation.
We convert MATLAB files into NumPy arrays to ensure reproducibility and faster experimentation in later stages.

```python
import os
import numpy as np
import scipy.io as sio

print("Working directory:", os.getcwd())
```

```python
data_path = "../data"
file_name = "AAS010R01.mat"
file_path = os.path.join(data_path, file_name)

print("File exists:", os.path.exists(file_path))

mat_data = sio.loadmat(file_path)
mat_data.keys()
```

```python
eeg = mat_data["signal"]           # (samples, channels)
stim_code = mat_data["StimulusCode"].flatten()
stim_type = mat_data["StimulusType"].flatten()

fs = 240  # sampling rate

print("EEG shape:", eeg.shape)
print("Stim code shape:", stim_code.shape)
print("Stim type shape:", stim_type.shape)
```

```python
np.save("../data/eeg.npy", eeg)
np.save("../data/stim_code.npy", stim_code)
np.save("../data/stim_type.npy", stim_type)

print("Saved numpy arrays successfully")
```
